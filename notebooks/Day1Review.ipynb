### The standard deviation

* Measures the "average" distance to the mean.
* Computation process
    * subtract the mean from each value (deviations)
    * square the deviations
    * sum the square deviations
    * compute the SD as follows
    
    $$ \sqrt{\frac{\text{total square deviations}}{n - 1}}$$

### Step-by-step computation in R

First, we compute the SD step-by-step for illustration.

Put data in a data frame and add a mean column

```r
df <- data.frame(x = xs)
df$mean = ave(df$x)

df
```

Subtract the mean from each value

```r
df$deviations <- df$x - df$mean
df
```

Square each deviation

```r
df$sqr_deviations <- df$deviations^2
df
```

Total the square deviations

```r
total_sqr_dev <- sum(df$sqr_deviations)
total_sqr_dev
```

Compute the SD from the formula
   $$ \sqrt{\frac{\text{total square deviations}}{n - 1}}$$

```r
n <- length(df$x)
sd <- sqrt(total_sqr_dev/(length(df$x) - 1))
sd
```

```r
# OR simply
sd(xs)
```

### Compute SD the fast way

* Normally, we don't want to do all that work.
* Use `sd` function

```r
sd(xs)
```

* Distribution of a statistic from many, many samples.

## Central Limit Theorem

* About the distribution of $\bar{x}$
* As sample size gets larger and larger, the distribution of $\bar{x}$ 
    * tends toward a bell curve
    * has a $mean(\bar{x}) = \mu$
    * has $sd(\bar{x}) = \frac{\sigma}{\sqrt(n)}$
