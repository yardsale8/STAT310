# Inverse Probability

## Finding the cut-off for a given tail percent

## Inverse Normal Problems

* Inverse Probability
    * Given a tail probability
    * Asked for the value/cutt-off
* Examples
    * What fraction of heights fall in the lowest 25% of the distribution?
    * What fraction of heights fall in the highest 30% of the distribution?


## Inverse Normal Probability on R

* Use `qnorm`
    * Default (`lower.tail = TRUE`) is left tail
    * Right tail use `lower.tail = FALSE`
    * Two tail find both tails
* Format is  `qnorm(proportion, mean, sd, lower.tail = TRUE)`
* Function for graphing is given below





```r
inverse_shade_normal <- function(p, mean, sd, lower.tail = TRUE) {
  l <- ifelse(lower.tail, mean-3.5*sd, qnorm(p, mean, sd, lower.tail = FALSE))
  u <- ifelse(lower.tail, qnorm(p, mean, sd), mean + 3.5*sd)
  cord.x <- c(l,seq(l,u,0.01),u) 
  cord.y <- c(0,dnorm(seq(l,u,0.01), mean, sd),0) 
  curve(dnorm(x,mean,sd),xlim=c(mean-3*sd,mean+3*sd),main='Normal Dist.') 
  polygon(cord.x,cord.y,col='skyblue')
}
```

## Example 2a

In Manoa Valley on the island of Oahu (Hawaii) the annual rainfall averages 43.6 inches with standard deviation 7.5 inches. A year is chosen at random. Assume a normal distribution.

**(a)** Find the cutoff for the largest 5% of years (in terms of annual rainfall). 


```r
inverse_shade_normal(0.05, 43.6, 7.5, lower.tail = FALSE)
```

```r
qnorm(0.05, 43.6, 7.5, lower.tail = FALSE)
```

## Example 2b

**(b)** Find the cutoff for the smallest 25% of years (in terms of annual rainfall). 

```r
inverse_shade_normal(0.25, 43.6, 7.5)
```

```r
qnorm(0.25, 43.6, 7.5)
```
