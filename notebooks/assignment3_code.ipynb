```r
n <- function(xs) {length(xs)}
df <- function(xs) {n(xs) - 1}
se <- function(xs) {sd(xs)/sqrt(n(xs))}
t_s <- function(xs, mu0) {(mean(xs) - mu0)/se(xs)}
pval <- function(xs) {pt(t_s(xs, 4), df(xs), lower.tail = FALSE)}
```

```r
xs <- rexp(10, 1/4)
pval(xs)
```

```r
replicate(10, pval(rexp(10, 1/4)))
```

```r
ps <- replicate(10000, pval(rexp(200, 1/4)))
```

```r
mean(ps < 0.05)
```
